```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
```

```python
n = 10
kappa = 3

graph = nx.random_regular_graph(kappa, n)
pos = nx.spring_layout(graph)
labels = np.random.choice([-1, 1], n)

ax = plt.subplot(111)
nx.draw_networkx(graph, pos=pos, arrows=False, labels=dict(enumerate(labels)),
                 node_size=1200, font_size=18)
ax.axis('off')
plt.savefig('../img/random_regular_graph.png', transparent=True)
```

```python
ax = plt.subplot(111)
nx.draw_networkx(graph, pos=pos, arrows=False, labels=dict(enumerate(labels)),
                 node_size=1200, font_size=18, node_color='#e86e78', edge_color='#ffffff', font_color='#ffffff')
ax.axis('off')
plt.savefig('../img/random_regular_graph.png', transparent=True)
```

```python
n = 4
m = 6
l = 3
k = 2

partite_graph = nx.algorithms.bipartite.configuration_model([l]*n, [k]*m)
sets = nx.bipartite.sets(partite_graph)
pos = nx.bipartite_layout(partite_graph, nodes=sets[0])
labels = np.random.choice([-1, 1], n + m)

nx.draw_networkx(partite_graph, pos=pos, arrows=False, labels=dict(enumerate(labels)))
# plt.savefig('../img/random_regular_graph.png')
```

```python
pos = nx.bipartite_layout(partite_graph, nodes=sets[0])
ax = plt.subplot(111)
nx.draw_networkx(partite_graph, pos=pos, arrows=False, labels=dict(enumerate(labels)),
                 node_size=1200, font_size=18, node_color='#e86e78', edge_color='#ffffff', font_color='#ffffff')
plt.xlim(-2, 1.7)
ax.axis('off')
plt.savefig('../img/random_regular_bipartite_graph.png', transparent=True)
```

```python
kappa = 4

graph = nx.star_graph(kappa)
# pos = nx.kamada_kawai_layout(graph)
x_pos = np.arange(kappa) - (kappa - 1) / 2
pos = np.vstack([[[0, 0]], np.vstack([x_pos, np.full_like(x_pos, -1)]).transpose()])
print(pos)
labels = np.random.choice([-1, 1], kappa + 1)

ax = plt.subplot(111)
ax.axis('off')
nx.draw_networkx(graph, pos=pos, arrows=False, labels=dict(enumerate(labels)),
                 node_size=1200, font_size=18, node_color='#e86e78', edge_color='#ffffff', font_color='#ffffff')
plt.axis('equal')
plt.savefig('../img/marked_star.png', transparent=True)
```
